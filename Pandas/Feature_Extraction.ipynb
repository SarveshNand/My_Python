```python
# loading the data
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r'anime.csv')
```

```python
df.head()
```

```python
# make a new column for episode count
```

```python
def extract_episodes(txt):
    check = False
    data = ""
    for i in txt:
        if i == ")":
            check = False
            return data
        if check == True:
            data = data + i
        if i == "(":
            check = True
```

```python
df['Episodes'] = df["Title"].apply(extract_episodes)
```

```python
df['Episodes'] = df['Episodes'].str.replace(" eps", "")
df['Episodes'] = df['Episodes'].astype(int)
```

```python
df
```

```python
# make a new column for time stamp
```

```python
df.loc[0]['Title']
```

```python
def extraction_time(txt):
    check = False
    data = ""
    for i in range(len(txt)):
        if txt[i] == ')':
            for j in range(i+1, i+20):
                data += txt[j]
            return data
```

```python
df['Total Time'] = df['Title'].apply(extraction_time)
```

```python
# which is the longest running anime
```

```python
df[['Start', 'End']] = df['Total Time'].str.split(' - ', expand=True)

df['Start'] = pd.to_datetime(df['Start'], format='%b %Y', errors='coerce')
df['End'] = pd.to_datetime(df['End'], format='%b %Y', errors='coerce')

df['End'] = df['End'].fillna(pd.Timestamp.today())

df['Duration_days'] = (df['End'] - df['Start']).dt.days
df['Duration_months'] = df['Duration_days'] / 30.44  # Approximate months
df['Duration_years'] = df['Duration_days'] / 365.25  # Approximate years

longest_running = df.sort_values(by='Duration_days', ascending=False).head(1)
print(longest_running[['Title', 'Start', 'End', 'Duration_days', 'Duration_years']])
```
