```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create a dataset with the following columns-> Products,Regions,Sales,Profit and 10 rows:
df = pd.DataFrame({
    'Products': ['Pen', 'Pencil', 'Eraser', 'Colour', 'Book', 'Notebook', 'Ink', 'Box', 'Cardboard', 'Paper'],
    'Regions': ['A', 'B', 'C', 'D', 'A', 'C', 'C', 'D', 'B', 'A'],
    'Sales': [20, 25, 27, 32, 23, 33, 37, 29, 30, 31],
    'Profit': [5, 3, 4, 7, 8, 9, 2, 6, 8, 9]
})
df
```

```python
# Plot a line chart of Sales for all 10 rows (index as X-axis).
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['Sales'], marker='o', linestyle='-', color='b')
plt.title('Sales Over Index')
plt.xlabel('Index')
plt.ylabel('Sales')
plt.grid(True)
plt.xticks(df.index)
plt.tight_layout()
plt.show()
```

```python
# Plot a bar chart showing total sales for each product.
plt.figure(figsize=(12, 6))
plt.bar(df['Products'], df['Sales'], color='skyblue')
plt.title('Total Sales per Product')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plot a scatter plot of Sales vs Profit.
plt.figure(figsize=(10, 5))
plt.scatter(df['Sales'], df['Profit'])
```

```python
# Create a pie chart showing total sales by region.
sales_by_region = df.groupby('Regions')['Sales'].sum()
plt.figure(figsize=(8, 8))
plt.pie(sales_by_region, labels=sales_by_region.index, autopct='%1.1f%%', startangle=140)
plt.title('Total Sales by Region')
plt.axis('equal')
plt.show()
```

```python
# Plot a histogram of the Sales column.
sns.histplot(df['Sales'])
```

```python
# Create a boxplot of Profit to detect outliers.
sns.boxplot(df['Profit'], palette='rainbow')
```

```python
# Create a countplot showing how many products are sold in each region.
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Regions', palette='pastel')
plt.title('Number of Products Sold per Region')
plt.xlabel('Region')
plt.ylabel('Number of Products')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Create a heatmap to show correlation between Sales and Profit.
corr_matrix = df[['Sales', 'Profit']].corr()
plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu', fmt='.2f', linewidths=0.5)
plt.title('Correlation between Sales and Profit')
plt.tight_layout()
plt.show()
```

```python
# Show total and average sales per region using pandas.
sales_summary = df.groupby('Regions')['Sales'].agg(['sum', 'mean']).rename(columns={'sum': 'Total Sales', 'mean': 'Average Sales'})
print(sales_summary)
```
